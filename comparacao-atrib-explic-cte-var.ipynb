```python
import numpy as np
from time import time
from matplotlib import pyplot as plt
```

```python
n = 10**6          # numero de operações
N = range(n)      
r = 500            # numero de repetições
R = range(r)
b = 1

taec = []
taev = []

sc = []
mc = []
vc = []

sv = []
mv = []
vv = []



for j in R:
    tic = time()
    for i in N:
        a = 1
    toc = time()
    
    taec.append(toc - tic)
    sc.append(np.std(taec))
    mc.append(np.mean(taec))
    vc.append(np.std(taec)/np.mean(taec))
    
for j in R:
    tic = time()
    for i in N:
        a = b
    toc = time()
    
    taev.append(toc - tic)
    sv.append(np.std(taev))
    mv.append(np.mean(taev))
    vv.append(np.std(taev)/np.mean(taev))
```

```python
#plotando resultados brutos

TAEC = np.mean(taec)
TAEV = np.mean(taev)

plt.plot(R,taec,'b.')
plt.plot(R,taev,'r.')

plt.legend(["atrib. expl. constante: Tmedio= {:.3g}".format(TAEC)+" seg",
            "atrib. expl. variavel: Tmedio= {:.3g}".format(TAEV)+" seg"])
plt.xlabel("repetição do experimento")
plt.ylabel("tempo de " +str(n)+ " atrib. explc.")
plt.grid()
plt.show()
```
