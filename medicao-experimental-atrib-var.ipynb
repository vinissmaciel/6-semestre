```python
#Adicionando repetições r


n = 10**6          # numero de operações
N = range(n)       # 
T = [] 
r = 1000           # numero de repetições
R = range(r)
b = 1

for j in R:
    tic = time()
    for i in N:
        a = b
    toc = time()
    
    #print(toc - tic)
    T.append(toc - tic)
```

```python
plt.bar(R, T)
plt.show()
```

```python
#plotando resultados brutos
plt.plot(R,T,'b.')
plt.legend(["atrib. expl. variavel"])
plt.xlabel("repetição do experimento")
plt.ylabel("tempo de " +str(n)+ " atrib. explc. var.")
plt.grid()
plt.show()
```

```python
#Calculando variabilidade para cada repetição

n = 10**6          # numero de operações
N = range(n)       # 
T = [] 
r = 1000           # numero de repetições
R = range(r)
V = []             # vetor de variabilidade
b = 1

for j in R:
    tic = time()
    for i in N:
        a = b
    toc = time()
    
    #print(toc - tic)
    T.append(toc - tic)
    
    S = np.std(T)
    M = np.mean(T)
    variabilidade = S/M
    V.append(variabilidade)
```

```python
plt.plot(V)
plt.show()
```

```python

n = 10**6          # numero de operações
N = range(n)       # 
T = [] 
r = 1000           # numero de repetições
R = range(r)
V = []             # vetor de variabilidade
S = []             # vetor de desvio padrao
M = []             # vetor de media
b = 1

for j in R:
    tic = time()
    for i in N:
        a = b
    toc = time()
    
    #print(toc - tic)
    T.append(toc - tic)
    
    S.append(np.std(T))
    M.append(np.mean(T))
```

```python
plt.plot(M)
plt.plot(S)
plt.grid()
plt.legend(["Média", "Desvio padrão"])
plt.show()
```
